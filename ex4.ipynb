```python
import numpy as np
from   unsupervised_learning import Pca, KernelPCA
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("Unsupervised_Learning_2024/Datasets/kPCA_data_2024.txt")
indices = np.random.choice(data.shape[0], 2000, replace=False)
data = data[indices, :]
labels = np.loadtxt("Unsupervised_Learning_2024/Datasets/kPCA_labels_2024.txt")
labels = labels[indices]
pca = Pca()
x = pca.fit_transform(data)
x = np.array(x)
plt.figure(figsize=(8,6))
plt.scatter(x[:,0],x[:,1], c=labels, cmap='viridis', s=10)
plt.show()
```

```python
kernel = KernelPCA('poly', degree=2)
y = kernel.fit_transform(data)
plt.figure(figsize=(8,6))
plt.scatter(y[:,0],y[:,1],c=labels, s=10)
plt.show()
```

```python
kernel = KernelPCA('poly', degree=5, gamma=2)
y = kernel.fit_transform(data)
plt.figure(figsize=(8,6))
plt.scatter(y[:,0],y[:,1],c=labels, s=10)
plt.show()
```

```python
kernel = KernelPCA('gauss',gamma=0.1)
y = kernel.fit_transform(data)
plt.figure(figsize=(8,6))
plt.scatter(y[:,0],y[:,1],c=labels, s=10)
plt.show()
```

```python
gaussk = KernelPCA('gauss',gamma=1)
y = gaussk.fit_transform(data)
plt.figure(figsize=(8,6))
plt.scatter(y[:,0],y[:,1],c=labels)
plt.show()
```

```python
kernel = KernelPCA('gauss',gamma=2)
y = kernel.fit_transform(data)
plt.figure(figsize=(8,6))
plt.scatter(y[:,0],y[:,1],c=labels)
plt.show()
```
