```python
import numpy as np
from   unsupervised_learning.PCA import Pca
import matplotlib.pyplot as plt
from typing import Literal
from  unsupervised_learning.KernelPCA import kernel_pca
```

```python
data = np.loadtxt("Unsupervised_Learning_2024/Datasets/kPCA_data_2024.txt")
labels = np.loadtxt("Unsupervised_Learning_2024/Datasets/kPCA_labels_2024.txt")
pca = Pca()
x = pca.fit_transform(data)
x = np.array(x)
plt.figure(figsize=(12,8))
plt.scatter(x[:,0],x[:,1], c=labels, cmap='viridis')
plt.plot()
```

```python
y = kpca.kernel_pca(data, 'polynomial')
plt.figure(figsize=(12,10))
plt.scatter(y[:,0],y[:,1],c=labels)
```
