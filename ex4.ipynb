```python
import numpy as np
from   unsupervised_learning import Pca, KernelPCA
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("ul_2024/Datasets/kPCA_data_2024.txt")
labels = np.loadtxt("ul_2024/Datasets/kPCA_labels_2024.txt")
pca = Pca()
x = pca.fit_transform(data)
x = np.array(x)
plt.figure(figsize=(8,6))
plt.scatter(x[:,0],x[:,1], c=labels, cmap='viridis', s=10)
plt.plot()
```

```python
kernel = KernelPCA('polynomial')
y = kernel.fit_transform(data)
plt.figure(figsize=(8,6))
plt.scatter(y[:,0],y[:,1],c=labels, s=10)
```

```python
kernel = KernelPCA('polynomial', delta=4)
y = kernel.fit_transform(data)
plt.figure(figsize=(8,6))
plt.scatter(y[:,0],y[:,1],c=labels, s=10)
```

```python
kernel = KernelPCA('gauss',3, 0.05)
y = kernel.fit_transform(data)
plt.figure(figsize=(8,6))
plt.scatter(y[:,0],y[:,1],c=labels, s=10)
```

```python
gaussk = KernelPCA('gauss',3, 1)
y = gaussk.fit_transform(data)
plt.figure(figsize=(8,6))
plt.scatter(y[:,0],y[:,1],c=labels, s=10)
```

```python
kernel = KernelPCA('gauss',3, 2)
y = kernel.fit_transform(data)
plt.figure(figsize=(8,6))
plt.scatter(y[:,0],y[:,1],c=labels, s=10)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(projection='3d')
ax.scatter(y[:,0],y[:,1],y[:,2], c=labels)
ax.view_init(10,10)
plt.show()
```

```python
import sklearn.decomposition as skpca
kernel = skpca.KernelPCA(2, kernel='linear', gamma=2)
y = kernel.fit_transform(data)
plt.figure(figsize=(8,6))
plt.scatter(y[:,0],y[:,1], c=labels)
plt.show()
```
