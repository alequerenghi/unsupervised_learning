```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
from unsupervised_learning.clustering import SpectralClustering
```

```python
iris = load_iris()
clusters = SpectralClustering(3, alg='ratio_cut').fit_predict(iris.data)
plt.scatter(iris.data[:,0],iris.data[:,1],c=clusters)
plt.show()
```
